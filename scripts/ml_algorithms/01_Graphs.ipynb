Importar bibliotecas

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Controle @16_Hz completo com FFT_64 

```python
# Pasta que contém os arquivos .npy
folder_path = r"C:\Users\Lenovo\Desktop\IC\[99] Database Final (16_0_fft_64)\cookie_c"

# Lista todos os arquivos .npy na pasta
npy_files = [f for f in os.listdir(folder_path) if f.endswith(".npy")]

# Criar uma figura para plotagem única
plt.figure(figsize=(10, 6))

# Usar uma paleta cíclica em vez de definir manualmente as cores
palette = sns.color_palette("Set2", n_colors=len(npy_files))

# Para visualizar todos os arquivos em um único gráfico, iterar sobre eles e plotar em uma única chamada
for idx, npy_file in enumerate(npy_files):
    # Carregue o arquivo .npy
    audio_data = np.load(os.path.join(folder_path, npy_file))
    
    # Plotar o áudio no gráfico, usando cores diferentes para cada arquivo
    plt.plot(audio_data, label=f"Áudio {idx+1} - {npy_file}", color=palette[idx % len(palette)])

# Adicionar rótulos e legenda
plt.title("Áudio de Vários Arquivos")
plt.xlabel("Amostras")
plt.ylabel("Amplitude")
plt.show()
```
